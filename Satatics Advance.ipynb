```python
Q1. What is the Probability Density Function (PDF)?
A Probability Density Function (PDF) is a function that describes the likelihood of a continuous random variable taking on a specific value. The area under the PDF curve between two points represents the probability that the variable falls within that range. The total area under the PDF curve equals 1.

    

Q2. What are the types of Probability Distribution?
The main types of probability distributions are:
1.	Normal Distribution: A symmetric, bell-shaped distribution where the mean, median, and mode coincide.
2.	Binomial Distribution: Describes the number of successes in a fixed number of Bernoulli trials.
3.	Poisson Distribution: Describes the probability of a given number of events occurring in a fixed interval of time or space.
4.	Uniform Distribution: All outcomes are equally likely within a certain range.
5.	Exponential Distribution: Describes the time between events in a Poisson process.

    
    
Q3. Write a Python function to calculate the probability density function of a normal distribution with a given mean and standard deviation at a given point.

import numpy as np

def normal_pdf(x, mean, std_dev):
    return (1 / (std_dev * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - mean) / std_dev) ** 2)

# Example usage
x = 1.5
mean = 0
std_dev = 1
pdf_value = normal_pdf(x, mean, std_dev)
pdf_value



Q4. What are the properties of Binomial Distribution? Give two examples of events where binomial distribution can be applied.
Properties of Binomial Distribution:
1.	Fixed number of trials (n): The process consists of a fixed number of independent trials.
2.	Two possible outcomes: Each trial has only two possible outcomes, success or failure.
3.	Constant probability of success (p): The probability of success is the same for every trial.
4.	Independent trials: The outcome of one trial does not affect the outcome of another.
Examples:
1.	Flipping a coin 10 times and counting the number of heads.
2.	Rolling a die 20 times and counting how many times a number greater than 4 appears.



Q5. Generate a random sample of size 1000 from a binomial distribution with a probability of success of 0.4 and plot a histogram of the results using matplotlib.

import numpy as np
import matplotlib.pyplot as plt

# Generate binomial distribution sample
n = 10  # Number of trials
p = 0.4  # Probability of success
sample = np.random.binomial(n, p, 1000)

# Plot histogram
plt.hist(sample, bins=10, edgecolor='black', alpha=0.7)
plt.title("Histogram of Binomial Distribution")
plt.xlabel("Number of Successes")
plt.ylabel("Frequency")
plt.show()




Q6. Write a Python function to calculate the cumulative distribution function of a Poisson distribution with a given mean at a given point.

import scipy.stats as stats

def poisson_cdf(k, mean):
    return stats.poisson.cdf(k, mean)

# Example usage
k = 3
mean = 5
cdf_value = poisson_cdf(k, mean)
cdf_value




Q7. How is Binomial Distribution different from Poisson Distribution?
•	Binomial Distribution: Deals with the number of successes in a fixed number of trials with a constant probability of success.
o	Example: Number of heads when flipping a coin 10 times.
•	Poisson Distribution: Describes the probability of a given number of events occurring in a fixed interval of time or space.
o	Example: The number of phone calls received at a call center in an hour.
The key differences are:
•	Binomial: Fixed number of trials, probability of success remains constant.
•	Poisson: No fixed number of trials, focuses on events occurring in a fixed interval.

    


Q8. Generate a random sample of size 1000 from a Poisson distribution with mean 5 and calculate the sample mean and variance.

import numpy as np

# Generate Poisson distribution sample
mean = 5
sample = np.random.poisson(mean, 1000)

# Calculate sample mean and variance
sample_mean = np.mean(sample)
sample_variance = np.var(sample)

sample_mean, sample_variance

               

Q9. How are mean and variance related in Binomial Distribution and Poisson Distribution?
•	Binomial Distribution:
o	Mean = n×pn \times pn×p (where nnn is the number of trials and ppp is the probability of success).
o	Variance = n×p×(1−p)n \times p \times (1 - p)n×p×(1−p).
•	Poisson Distribution:
o	Mean = λ\lambdaλ (rate parameter).
o	Variance = λ\lambdaλ.
Thus, for a Poisson distribution, the mean and variance are equal.

            

Q10. In normal distribution with respect to mean position, where does the least frequent data appear?
In a normal distribution, the least frequent data appears in the tails of the distribution, far from the mean. These points are more than 3 standard deviations away from the mean.
4o

```
